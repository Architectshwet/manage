```python
import re
import csv

def parse_questions_to_csv(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        content = f.read()

    # Split the file content into blocks based on "Question <number>"
    blocks = re.split(r'^Question \d+\s*', content, flags=re.MULTILINE)
    
    data = []
    for block in blocks:
        # Skip empty blocks or any headers before the first question
        if not block.strip() or 'Answer:' not in block:
            continue
            
        # Extract Question Text (everything up to "A. " at the start of a line)
        question_match = re.search(r'^\s*(.*?)(?=^A\.)', block, re.MULTILINE | re.DOTALL)
        
        # Extract Options
        opt_a_match = re.search(r'^A\.\s*(.*?)(?=^B\.)', block, re.MULTILINE | re.DOTALL)
        opt_b_match = re.search(r'^B\.\s*(.*?)(?=^C\.)', block, re.MULTILINE | re.DOTALL)
        opt_c_match = re.search(r'^C\.\s*(.*?)(?=^D\.)', block, re.MULTILINE | re.DOTALL)
        opt_d_match = re.search(r'^D\.\s*(.*?)(?=^E\.|^F\.|^Answer:)', block, re.MULTILINE | re.DOTALL)
        opt_e_match = re.search(r'^E\.\s*(.*?)(?=^F\.|^Answer:)', block, re.MULTILINE | re.DOTALL)
        opt_f_match = re.search(r'^F\.\s*(.*?)(?=^Answer:)', block, re.MULTILINE | re.DOTALL)
        
        # Extract Answer
        ans_match = re.search(r'^Answer:\s*(.*)', block, re.MULTILINE)
        
        # Clean extracted strings
        question = question_match.group(1).strip() if question_match else ""
        opt_a = opt_a_match.group(1).strip() if opt_a_match else ""
        opt_b = opt_b_match.group(1).strip() if opt_b_match else ""
        opt_c = opt_c_match.group(1).strip() if opt_c_match else ""
        opt_d = opt_d_match.group(1).strip() if opt_d_match else ""
        opt_e = opt_e_match.group(1).strip() if opt_e_match else ""
        opt_f = opt_f_match.group(1).strip() if opt_f_match else ""
        ans = ans_match.group(1).strip() if ans_match else ""
        
        data.append({
            'Question': question,
            'Option A': opt_a,
            'Option B': opt_b,
            'Option C': opt_c,
            'Option D': opt_d,
            'Option E': opt_e,
            'Option F': opt_f,
            'Answer': ans,
            'Times Asked': ""
        })

    # Define the required CSV headers
    fieldnames = [
        'Question', 
        'Option A', 
        'Option B', 
        'Option C', 
        'Option D', 
        'Option E', 
        'Option F', 
        'Answer', 
        'Times Asked'
    ]
    
    # Write to CSV
    with open(output_file, 'w', newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)
        
    print(f"Successfully processed {len(data)} questions and saved to {output_file}")

if __name__ == "__main__":
    parse_questions_to_csv('databricks_generative_ai/6.txt', 
                           'databricks_generative_ai/6.csv')
```
